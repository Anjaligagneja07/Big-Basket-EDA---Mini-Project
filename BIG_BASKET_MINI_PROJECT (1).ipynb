### **STEP 5: TOP & LEAST SOLD PRODUCTS**
**Note: The dataset does not include 'quantity sold'. So i have assumed ratings as a proxy with sale price.**

```python
top_rated = df.sort_values(by='rating', ascending=False).head(10)
least_rated = df.sort_values(by='rating', ascending=True).head(10)

print("\nTop 10 Rated Products (with Sale Price):")
print(top_rated[['product', 'rating', 'sale_price']])

print("\nLeast 10 Rated Products (with Sale Price):")
print(least_rated[['product', 'rating', 'sale_price']])
```

### **STEP 6: MEASURING DISCOUNT ON A CERTAIN ITEM.**

```python
# Discount = market_price - sale_price
df['discount'] = df['market_price'] - df['sale_price']
df['discount_percent'] = (df['discount'] / df['market_price']) * 100

print("Sample Discounts:")
df[['product', 'sale_price', 'market_price', 'discount', 'discount_percent']].head()
```

### **STEP 7: MISSING VALUES**

```python
# finding missing values if any
print("Missing Values in Dataset:")
df.isnull().sum()
```

### **STEP 8: FIND OUTLIERS AND FILL THEM WITH MEAN.**

```python
# handling outliers
for column in ['sale_price', 'market_price', 'rating']:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Outliers in {column}: {len(outliers)}")
    mean_val = df[column].mean()
    df.loc[(df[column] < lower_bound) | (df[column] > upper_bound), column] = mean_val
```

### **STEP 9: VISUALIZATIONS**

### **1. DISTRIBUTION OF CATEGORIES**

```python
plt.figure(figsize=(10, 6))
df['category'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Product Categories', fontweight = 'bold', color = 'blue')
plt.xlabel('Category', fontweight = 'bold', color = 'blue')
plt.ylabel('Count', fontweight = 'bold', color = 'blue')
plt.tight_layout()
plt.show()
```

**Most products belong to a few dominant categories, indicating a concentration in specific product types.**

### **2. TOP BRANDS BY PRODUCT COUNT**

```python
plt.figure(figsize=(10, 6))
df['brand'].value_counts().head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Brands by Product Count',fontweight = 'bold', color = 'green')
plt.xlabel('Brand', fontweight = 'bold', color = 'green')
plt.ylabel('Count', fontweight = 'bold', color = 'green')
plt.tight_layout()
plt.show()
```

**The top 10 brands contribute significantly to the product range, highlighting brand dominance in the dataset.**

### **3. RATINGS DISTRIBUTION**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'],bins=20, kde=True, color='green')
plt.title('Ratings Distribution', fontweight = 'bold')
plt.xlabel('Rating', fontweight = 'bold')
plt.ylabel('Frequency', fontweight = 'bold')
plt.tight_layout()
plt.show()
```

**Ratings are skewed towards higher values, indicating overall positive customer feedback.**

### **4. DISCOUNT PERCENT BY CATEGORY**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='category', y='discount_percent',data=df)
plt.title('Discount Percent by Category', fontweight ='bold', color = 'red')
plt.xlabel('category', fontweight = 'bold', color = 'red')
plt.ylabel('discount_percent', fontweight = 'bold', color = 'red')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Discount percentages vary significantly across categories, with some categories offering notably higher discounts than others.**

### **5. CORRELATION HEATMAP**

```python
plt.figure(figsize=(40, 25))
sns.heatmap(df[['sale_price', 'market_price', 'rating', 'discount', 'discount_percent']].corr(), annot=True, annot_kws={"size": 40},
            cmap='coolwarm')
plt.title('Correlation Heatmap', color = 'orange',fontweight = 'bold', fontsize = 45)
plt.xticks(fontsize=35,fontweight = 'bold')
plt.yticks(fontsize=35,fontweight = 'bold')
plt.tight_layout()
plt.show()
```

**Sale price is highly correlated with market price and discount values, indicating pricing strategies are closely linked to discounts.**
